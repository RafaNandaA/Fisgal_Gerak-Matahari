```python
import astropy.io.fits as fits
import pandas as pd

fits_file = r'J_A+A_530_A138_catalog.dat.gz\J_A+A_530_A138_catalog.dat.gz.fits'
hdul = fits.open(fits_file)
hdul.info()
data = hdul[1].data

df = pd.DataFrame(data)

hdul.close()

print(df)
```

```python
print(df.head())
print(df.dtypes)
```

```python
print(df.columns)
```

```python
f_df = df[(df['Teff'] > 5500) & (df['Teff'] < 7200)& (df['logg'] <= 5)& (df['logg'] >=4)& (df['UVel'] >=-10000)]

```

```python
import matplotlib.pyplot as plt
U = f_df['UVel']
V = f_df['VVel']
W = f_df['WVel']

plt.hist(U,100)
plt.show()
plt.hist(V,100)
plt.show()
plt.hist(W,100)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
data = U
bin = int(np.sqrt(len(data)))
plt.hist(data, bins=bin, density=True, alpha=0.6, color='g', label='Data Histogram')
mean, std_dev = norm.fit(data)
def gaussian(x, mean, std_dev):
    return 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std_dev**2))
popt, _ = curve_fit(gaussian, data, norm.pdf(data, mean, std_dev), p0=[mean, std_dev])
plt.plot(data, gaussian(data, *popt), color='r', label='Fitted Gaussian Curve')
plt.text(0.5, 0.5, f'Mean: {mean:.2f}\nStd Dev: {std_dev:.2f}', transform=plt.gca().transAxes)
plt.legend()
plt.show()
```
